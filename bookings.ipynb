Неудобно работать с переменным где названия в несколько слов и с заглавными буквами, поправим это.

```python
def replace(name):
    new_name = name.replace(' ', '_')
    return new_name
```

Необходимо найти топ стран по успешным бронирояниям (**is_canceled == 0**)

```python
bookings.query('is_canceled == 0')\
    .country.value_counts()[:5]
```

Менеджерам необходима информация на сколько ночей  в среднем бронируют отели типа City Hotel? Resort Hotel?  

Данные необходимы  простой десятичной дроби.

```python
bookings.groupby('hotel', as_index=False)\
    .aggregate({'stays_total_nights' :'mean' })\
    .round(2)\
    .rename({'stays_total_nights': 'mean_days'})
```

```python
bookings.query('is_canceled == 1')\
    .value_counts('country').sum()
```

Иногда тип номера, полученного клиентом , отличается от изначально забронированного . Такое может произойти, например, по причине овербукинга.
Сколько подобных наблюдений встретилось в датасете?

```python
bookings.query('assigned_room_type != reserved_room_type').shape
```

На какой месяц чаще всего оформляли бронь в 2016 году? Изменился ли самый популярный месяц в 2017 году?

```python
bookings.groupby(['arrival_date_year','arrival_date_month'], as_index=False)\
        .aggregate({'hotel' : 'count'})\
        .sort_values(by='hotel', ascending=False)[:5]
```

На какой месяц (arrival_date_month) бронирования отеля типа City Hotel отменялись чаще всего в 2015, 2016, 2017 годах? 

```python
bookings.query('is_canceled == 1 and hotel =="City Hotel" ')\
        .groupby('arrival_date_year')\
        .arrival_date_month\
        .value_counts()
        
```

Логичнее всего продположить что отели бронируют взрослые, путешествуя одни или с детьми, проверим так ли это.

```python
bookings[['adults', 'children' , 'babies']].mean()
```

Определимся каких отелей бронируют больше с детми (городских или курортных).

```python
bookings['total_kids']= bookings['children'] + bookings['babies']
bookings.groupby('hotel', as_index=False)\
        .aggregate({'total_kids' : 'mean'})\
        .round(2)\
        .rename({'total_kids': 'avg_kigs'})
```

Не все бронирования завершились успешно, поэтому посчитаем сколько клиентов было потеряно в процессе, связоно ли это с наличием детей?

```python
bookings['has_kids']=bookings.total_kids > 0
yes_kids_churn = round(bookings.query('is_canceled == 1 and has_kids == True')
        .shape[0] /
bookings.query('has_kids == True')
        .shape[0] *100, 2)
yes_kids_churn
```

```python
bookings['has_kids']=bookings.total_kids > 0
no_kids_churn = round(bookings.query('is_canceled == 1 and has_kids == False')
        .shape[0] /
bookings.query('has_kids == False')
        .shape[0] *100, 2)
no_kids_churn
```
